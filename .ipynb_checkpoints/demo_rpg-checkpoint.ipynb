# 🎮 Démo RPG
Démonstration interactive de toutes les features du projet.

| Feature | Statut |
|---|---|
| Endurance / Niveau / Force | ✅ Prête |
| Dégâts aléatoires (0 → D) | ✅ Prête |
| Duel 2v2 | ✅ Prête |

```python
import sys
sys.path.insert(0, 'src')

from rpg.character import Character
from rpg.team import Team, duel

print('✅ Imports OK')
```

## ⚔️ Feature 1 — Statistiques : Endurance, Niveau, Force
- HP = 10 + END
- HP et dégâts augmentent de 2 par niveau
- Dégâts = 1 + FOR + 2×niveau

```python
# Personnage de base
alice = Character('Alice')
print(f'Alice (base)        → HP: {alice.health}, Dégâts max: {1 + alice.force + 2 * alice.level}')

# Avec endurance
tank = Character('Tank', endurance=5)
print(f'Tank (END=5)        → HP: {tank.health}, Dégâts max: {1 + tank.force + 2 * tank.level}')

# Avec niveau
veteran = Character('Vétéran', level=3)
print(f'Vétéran (LVL=3)     → HP: {veteran.health}, Dégâts max: {1 + veteran.force + 2 * veteran.level}')

# Avec force
warrior = Character('Guerrier', force=4)
print(f'Guerrier (FOR=4)    → HP: {warrior.health}, Dégâts max: {1 + warrior.force + 2 * warrior.level}')

# Combiné
hero = Character('Héros', endurance=3, level=2, force=2)
print(f'Héros (END=3,LVL=2,FOR=2) → HP: {hero.health}, Dégâts max: {1 + hero.force + 2 * hero.level}')
```

## 🎲 Feature 2 — Dégâts aléatoires entre 0 et D
À chaque attaque, les dégâts sont tirés aléatoirement entre 0 et D.

```python
attacker = Character('Attaquant', force=3, level=1)
d_max = 1 + attacker.force + 2 * attacker.level
print(f'Attaquant (FOR=3, LVL=1) → D max = {d_max}')
print()

print('Simulation de 10 attaques :')
for i in range(10):
    target = Character('Cible')
    hp_before = target.health
    attacker.attack(target)
    degats = hp_before - target.health
    print(f'  Attaque {i+1:2d} → dégâts: {degats} (HP cible: {target.health})')
```

## 🏆 Feature 3 — Duel 2v2
Deux équipes de deux personnages s'affrontent. Chaque vivant attaque un ennemi vivant aléatoire. Dernier survivant gagne.

```python
team_a = Team('Équipe A', [
    Character('Alice', endurance=2, level=1, force=1),
    Character('Bob',   endurance=0, level=2, force=2),
])

team_b = Team('Équipe B', [
    Character('Charlie', endurance=3, level=1, force=0),
    Character('Dave',    endurance=1, level=0, force=3),
])

print('=== Avant le duel ===')
for m in team_a.members + team_b.members:
    print(f'  {m.name:10s} HP: {m.health}')

winner = duel(team_a, team_b)

print()
print('=== Après le duel ===')
for m in team_a.members + team_b.members:
    statut = '💀 mort' if m.is_dead() else f'❤️  {m.health} HP restants'
    print(f'  {m.name:10s} {statut}')

print()
print(f'🏆 Vainqueur : {winner.name}')
```

## 🛡️ Feature 4 — Score d'armure selon l'équipement
> 🔧 **Stub** — remplacer par `from rpg.armor import Armor` quand la feature est prête.

```python
# ---- STUB : remplacer ce bloc par le vrai import quand prêt ----
class Armor:
    """STUB - Score d'armure selon l'équipement"""
    def __init__(self, name: str, defense: int):
        self.name = name
        self.defense = defense

    def __repr__(self):
        return f'Armor({self.name}, DEF={self.defense})'
# ----------------------------------------------------------------

leather = Armor('Cuir',  defense=2)
chain   = Armor('Cotte', defense=5)
plate   = Armor('Plaques', defense=8)

print('Équipements disponibles :')
for armor in [leather, chain, plate]:
    print(f'  {armor.name:10s} → DEF: {armor.defense}')

# Exemple d'utilisation anticipée
hero = Character('Héros', endurance=2)
hero.armor = plate  # sera géré par Character quand la feature sera prête
print(f'\n{hero.name} équipé de {hero.armor.name} → DEF: {hero.armor.defense}')
```

## 🗡️ Feature 5 — Dégâts selon l'arme
> 🔧 **Stub** — remplacer par `from rpg.weapon import Weapon` quand la feature est prête.

```python
# ---- STUB : remplacer ce bloc par le vrai import quand prêt ----
class Weapon:
    """STUB - Dégâts selon l'arme"""
    def __init__(self, name: str, damage_bonus: int):
        self.name = name
        self.damage_bonus = damage_bonus

    def __repr__(self):
        return f'Weapon({self.name}, +{self.damage_bonus} dégâts)'
# ----------------------------------------------------------------

dagger = Weapon('Dague',  damage_bonus=1)
sword  = Weapon('Épée',   damage_bonus=3)
axe    = Weapon('Hache',  damage_bonus=5)

print('Armes disponibles :')
for w in [dagger, sword, axe]:
    print(f'  {w.name:10s} → +{w.damage_bonus} dégâts')

# Exemple d'utilisation anticipée
hero = Character('Héros', force=2)
hero.weapon = sword  # sera géré par Character quand la feature sera prête
d_base = 1 + hero.force + 2 * hero.level
d_avec_arme = d_base + hero.weapon.damage_bonus
print(f'\n{hero.name} avec {hero.weapon.name} → D max: {d_base} + {hero.weapon.damage_bonus} = {d_avec_arme}')
```

## 🎬 Scénario complet — Tout en action
Un duel 2v2 avec toutes les features actives (les stubs seront remplacés au fur et à mesure).

```python
print('=== DUEL FINAL ===')
print()

team1 = Team('Les Héros', [
    Character('Arthur', endurance=4, level=2, force=3),
    Character('Merlin', endurance=1, level=3, force=1),
])

team2 = Team('Les Vilains', [
    Character('Mordred', endurance=2, level=2, force=4),
    Character('Morgan',  endurance=3, level=1, force=2),
])

print('Composition des équipes :')
for team in [team1, team2]:
    print(f'\n  {team.name}')
    for m in team.members:
        d = 1 + m.force + 2 * m.level
        print(f'    {m.name:10s} HP: {m.health:3d} | D max: {d}')

print()
winner = duel(team1, team2)

print('Résultat :')
for team in [team1, team2]:
    for m in team.members:
        statut = '💀' if m.is_dead() else f'❤️  {m.health} HP'
        print(f'  {m.name:10s} {statut}')

print(f'\n🏆 Victoire : {winner.name} !')
```
